```python
# IMPORTING SOFTWARE AND CSV FILE   
    
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt 
clgd = pd.read_csv(r'C:\Users\knottszd\Desktop\Cell_Line_and_Gene_Data.csv', index_col=0)
clgd_adj = clgd.drop(clgd.index[0:60])
clgd_fadj = clgd_adj.drop(clgd.index[41248:])
    
    
#DEFINING ZK1 and ZK2, PEARSONS CORRELATION COEFFICIENT    
    
def zk1(x1, y1, genename1, genename2, a):
    '''Math for Computing A, B, R, R^2 for Geneanalysis'''
    ey = sum(y1)
    ey2 = sum(y1) ** 2
    ey3 = sum(y1 ** 2)
    n = 76
    ex = sum(x1)
    ex2 = sum(x1) ** 2    
    ex3 = sum(x1 ** 2)
    exy = sum(x1 * y1)
    A = ((ey * ex3) - (ex * exy)) / ((n * ex3) - (ex2))
    B = ((n * exy) - (ex * ey)) / ((n * ex3) - ex2)
        
#PRINTING DATA AND MATPLOTLIB GRAPHS

    print('Statistical Regression Test:')
    print("Y-Intercept:", A, "Slope:", B)
    r = ((n * exy) - (ex * ey)) / ((((n * ex3) - (ex2)) * ((n * ey3) - (ey2))) ** (1 / 2))
    R = round(r, 3)
    r2 = r ** 2
    R2 = round(r2, 3)
    print("R^2 = ", R2, "R = ", R)
    if r2 <= .7:
        print('Not significant')
    elif r2 > .7:
        print("Significant")
            
#Matplotlib Graph Functions
    x = plt.figure()
    plt.scatter(x1, y1)
    plt.xlabel(str(genename1) + " Gene Expression")
    plt.ylabel(str(genename2) + " Gene Expression")
    plt.title(str(genename1) + ' vs ' + str(genename2))
    plt.grid(True)
    plt.show()
        
        
#SAVING MATPLOTLIB GRAPH AS A PDF FILE
        
    x.savefig('test' + str(genename1) + '.pdf')
        
        
def zk2(x1, y1, genename1, genename2, a):
    '''Math for Computing A, B, R, R^2 for Multigene'''
    ey = sum(y1)
    ey2 = sum(y1) ** 2
    ey3 = sum(y1 ** 2)
    n = 76
    ex = sum(x1)
    ex2 = sum(x1) ** 2
    ex3 = sum(x1 ** 2)
    exy = sum(x1 * y1)
    a = ((ey * ex3) - (ex * exy)) / ((n * ex3) - (ex2))
    A = round(a, 3)
    b = ((n * exy) - (ex * ey)) / ((n * ex3) - ex2)
    B = round(b, 3)
    r = ((n * exy) - (ex * ey)) / ((((n * ex3) - (ex2)) * ((n * ey3) - (ey2))) ** (1 / 2))
    R = round(r, 3)
    r2 = r ** 2 
    R2 = round(r2, 3)

#MATPLOTLIB GRAPH FUNCTIONS WITH IF STATEMENTS

    if r >= .7:
        print('Statistical Regression Test for ' + str(genename1) + ':')
        print("Y-Intercept:", A, "Slope:", B)
        print("R^2= ", R2, "R= ", R)
        print("Significant")
        x = plt.figure()
        par = np.polyfit(x1, y1, 1, full=True)
        slope=par[0][0]
        intercept=par[0][1]
        xl = [min(x1), max(x1)]
        yl = [slope*xx + intercept  for xx in xl]
        plt.plot(xl, yl, '-r')
        plt.scatter(x1, y1)
        plt.xlabel(str(genename1) + " Gene Expression")
        plt.ylabel(str(genename2) + " Gene Expression")
        plt.title(str(genename1) + ' vs ' + str(genename2))
        plt.text((min(x1)), (max(y1)-1), 'R ='),
        plt.text((min(x1)), (max(y1)-1.5), str(R))
        plt.grid(True)
        plt.show()
        x.savefig('test' + str(genename1) +'vs'+ str(genename2) + '.pdf')
    return r
    if r <= -.7:
        print('Statistical Regression Test for ' + str(genename1) + ':')
        print("Y-Intercept:", A, "Slope:", B)
        print("R^2= ", R2, "R= ", R)
        print("Significant")
        x = plt.figure()
        par = np.polyfit(x1, y1, 1, full=True)
        slope=par[0][0]
        intercept=par[0][1]
        xl = [min(x1), max(x1)]
        yl = [slope*xx + intercept  for xx in xl]
        plt.plot(xl, yl, '-r')
        plt.scatter(x1, y1)
        plt.xlabel(str(genename1) + " Gene Expression")
        plt.ylabel(str(genename2) + " Gene Expression")
        plt.title(str(genename1) + ' vs ' + str(genename2))
        plt.text((max(x1)-1), (max(y1)-1), 'R ='),
        plt.text((max(x1)-1), (max(y1)-1.5), str(R))
        plt.grid(True)
        plt.show()
        x.savefig('test' + str(genename1) +'vs'+ str(genename2) + '.pdf')
    return r
        
    
   
def Geneanalysis():
    """GENE ONE COMPARISON TO GENE 2"""
        
#INPUT BOTH GENES INTO THE INDEX USING LOC
    gene1 = input("Enter Gene 1:")
    gene2 = input("Enter Gene 2:")
    x1 = clgd_fadj.loc[gene1]
    y1 = clgd_fadj.loc[gene2]
        
 
    if type(x1) == pd.core.frame.DataFrame:
#IF X1 IS A DATAFRAME             
            
        for index, row in x1.iterrows():
            if (type(y1) == pd.core.frame.DataFrame):
                # IF A GENE TYPE HAS MORE THAN 1 PROBE/ROW
                a = 0
                    
                for ind, rw in y1.iterrows():
                    # ITERATE THROUGH THE ROWS AND COMPUTE VALUES FOR EACH COMBO
                    zk1(row, rw, gene1, gene2, a)
                    a +=1
            else:
                # IF THERE IS ONLY ONE ROW FOR A GENE TYPE
                zk1(row, y1, gene1, gene2, a)
                
    else:
#IF X1 IS ONLY ONE ROW 
            
        if (type(y1) == pd.core.frame.DataFrame):
            #IF A GENE TYPE HAS MORE THAN ONE ROW 
            a = 0
                
            for ind, rw in y1.iterrows():
                #ITERATE THROUGH THE ROWS AND COMPUTE VALUES FOR EACH COMBO 
                zk1(x1, rw, gene1, gene2, a)
                a += 1
        else:
            # IF THERE IS ONLY ONE ROW FOR A GENE TYPE 
            zk1(x1, y1, gene1, gene2, a)

    
    
    
def Multigene():
    """ONE GENE COMPARED TO EVERY GENE"""

#INPUT BOTH GENES INTO TEH INDEX USING LOC 
    gene1 = input("Enter Gene:")
    allgenes = clgd_fadj.index[0:41248]
        
    #CREATE LIST FOR SIGNIFICANT GENES
    significant_genes = []        
        
    #X1 IS THE INPUT, Y1 IS THE ENTIRE INDEX
    x1 = clgd_fadj.loc[gene1]
    y1 = clgd_fadj.loc[allgenes]
        
        
    if type(x1) == pd.core.frame.DataFrame:
#IF X1 IS A DATAFRAME        
            
        for index, row in x1.iterrows():
            #FOR THE COLUMNS AND ROWS IN X1
                
            if (type(y1) == pd.core.frame.DataFrame):
                # IF Y1 HAS MORE THAN ONE ROW 
                a=0
                    
                for ind, rw in y1.iterrows():
                    # USE IND TO FIND CORRESPONDING GENE NAME 
                    # ITERATE THROUGH THE ROWS AND COMPUTE VALUES FOR EACH COMBO 
                    r = zk2(row, rw, gene1, ind, a)
                    a+=1
                    if r > 0.7 or r < -0.7:
                        significant_genes.append(ind) 
                    #print(y1.loc[rw])
                
            else:
                # IF THERE IS ONLY ONE ROW FOR A GENE TYPE 
                r = zk2(row, y1, gene1, ind, a)
                if r > 0.7 or r < -0.7:
                    significant_genes.append(ind) 
        
    else:
# IF X1 IS ONLY ONE ROW 
            
        if (type(y1) == pd.core.frame.DataFrame):
            # IF A GENE TYPE HAS MORE THAN ONE ROW 
                
            for ind, rw in y1.iterrows():
                a=0
                #ITERATE THROUGHT THE ROWS AND COMPUTE VALUES FOR EACH COMBO 
                r = zk2(x1, rw, gene1, ind, a)
                a+=1
                if r > 0.7 or r < -0.7:
                    significant_genes.append(ind)             
            
        else:
            # IF THERE IS ONLY ONE ROW FOR A GENE TYPE 
            r = zk2(x1, y1, gene1, ind, a)
            if r > 0.7 or r < -0.7:
                    significant_genes.append(ind) 
        
    print(significant_genes)
        
        
     
                                
```
